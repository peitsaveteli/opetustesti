Tässä datassa tarkastellaan samasta ajosta syntyneitä [myonipareja](https://fi.wikipedia.org/wiki/Myoni), joilla on haluttuja ominaisuuksia (kuten vastakkaiset sähkövaraukset). Kunkin ajon aikana kiihdyttimessä ajetaan vastakkaisiin suuntiin joukkoja hiukkasia, jotka sitten törmäilevät mitta-asemien kohdalla ja näistä törmäyksistä sinkoilevia uusia hiukkasia voidaan havainnoida.

Mutta! Se mistä usein ollaan oikeasti kiinnostuneita ei välttämättä ole suoraan nähtävissä. Teorioiden mukaan törmäyksien tuottamissa olosuhteissa voisi esiintyä kiinnostavia hiukkasia, jotka ovat kuitenkin niin lyhytikäisiä ettei niitä koskaan ehdi nähdä itsenään ennen kuin ne ovat jo tuhoutuneet. Tämä tuhoutuminen kuitenkin tapahtuu niin, että hiukkanen hajoaa kevyemmiksi yksilöiksi ja niitä havainnoidaan itse koelaitteistolla. Tällainen *epäsuora havainnointi* on luonnontieteissä yllättävän yleistä.

Tuhoutuminen on oikeastaan aika huono sana, sillä luonnossa asiat tuppaavat säilymään, joskin muotoaan muuttaneina. Nämä hajoamisreaktioissa etsityt hiukkasetkin voidaan päätellä tiettyjen *säilymislakien* avulla. Tässä yhteydessä kyse on siitä, että suhteellisuusteoria huomioiden tiedetään syntyneiden hiukkasten energioiden ja liikkeiden vastaavan tietyn massaista isompaa hiukkasta. Toisin sanoen jos tiedetään millaisia hiukkasia hajoamistilanteesta on lähtenyt, minne ne ovat menneet ja millä nopeuksilla, voidaan arvioida mistä hiukkasesta ne ovat syntyneet.

### 5. Kadonneen hiukkasen metsästys

Aiemmin mainittu säilyvä asia on niin kutsuttu *invariantti massa*. Se voidaan laskea käsitellystä datasta seuraavasti:

```python
# Luodaan muuttuja 'invariantti_massa' joka kertoo kunkin 'datasetti'-muuttujan rivin mukaisen inv. massan.
# Tässä siis lasketaan hitusen lukiomatikkaa vaativammalla tasolla edellisen taulukon osista uutta tietoa.

invariantti_massa = np.sqrt(2*datasetti.pt1*datasetti.pt2*(np.cosh(datasetti.eta1-datasetti.eta2) - np.cos(datasetti.phi1-datasetti.phi2)))
```

```python
# Katsotaan miltä se näyttää.

invariantti_massa.head()
```

```python
# Katsotaan nyt sen pituuskin, että ollaan varmasti saatu arvo joka riville.

len(invariantti_massa)
```

Nyt meillä on koko 'datasetti'-muuttujaa pienempi palanen, yksi siitä laskettu lisäsarake, tarkemmassa syynissämme. Alun ikäesimerkin tapaan voimme nyt tehdä tästä invariantin massan listasta histogrammin, joka kertoo miten eri tapauksista lasketut massa-arviot ovat jakautuneet. Koska hiukkaset ovat arkikokemukseen nähden niin kevyitä, että kilogrammojen käyttö olisi hankalasti esitettävää, käytetään massan yksikkönä Gev/c$^2$ (gigaelektronivoltteja per valonnopeuden neliö, syy löytyy energian ja massan suhteesta E=mc$^2$). Välillä c$^2$ vain pudotetaan pois, koska on selvää mistä puhutaan.

```python
# Suoritetaan histogrammin piirtäminen:

plt.hist(invariantti_massa, bins=100, range=(80,100))

# Näillä riveillä ainoastaan määritellään otsikko sekä akseleiden tekstit.

plt.xlabel('Invariantti massa [GeV/c²]')
plt.ylabel('Tapahtumien lukumäärä')
plt.title('Kahden myonin invariantin massan histogrammi \n') # \n luo uuden rivin otsikon muotoilua varten

# Tehdään kuvaaja näkyväksi.

plt.show()

# Aja koodi useampaan kertaan:
# Mitä tapahtuu, jos muutat binien määrää? Mitä bins määrittää? Mikä on mielestäsi paras arvo?
# Entä range? Mitä käy, jos otat rangen määritelmän kokonaan pois? 
```

Mitä tulos mielestäsi osoittaa? Miten luotettavaksi arvioisit näkemäsi?

[Z-bosonin](https://fi.wikipedia.org/wiki/W-_ja_Z-bosonit), raskaan ja hajoavan hiukkasen, massa on 91,1876 ± 0,0021 GeV/c$^2$. Uskoisitko kuvaajan perusteella, että sellainen saattaisi löytyä hajoamistapahtumiemme takaa?


### 6. Mittapisteiden vaikutuksesta

Kokeillaan sitten samaa, mutta niin että vaihdellaan tarkasteltavien mittauspisteiden määrää. Tieteenteossa yleinen nyrkkisääntö on, että mitä enemmän mittapisteitä otat, sitä paremmin oikeasti luonnossa vallitsevat riippuvaisuudet alkavat nousemaan esiin taustakohinan seasta.

Seuraava koodisolu voi näyttää hankalalta, mutta siitä ei kannata ottaa stressiä. Tärkeintä on ajaa sitä ja ohjelman kysyessä syötteeksi lukua nollan ja datassa olevien mittapisteiden määrän välillä antaa sille luku. Aja solua useamman kerran eri arvoilla ja pohdi samalla, mikä syntyvistä kuvaajista vaikuttaa luotettavimmalta ja miksi.

```python
# Luodaan tyhjä taulukko 'valitut', johon tullaan tallentamaan valittu määrä invariantin massan arvoja.
# Tietokone on vähän tällainen, sille pitää kertoa että jokin on olemassa tai se ei usko.

valitut = []

# Pyydetään käyttäjältä tapahtumien määrä ja tallennetaan sen muuttujaan 'lkm'.
# Koodi vaatii toimiakseen kokonaislukusyötteen (integer)

lkm = int(input('Syötä haluttu tapahtumien määrä: '))

# Luodaan if-silmukka, joka tutkii syötettyä tapahtumien määrää eli muuttujan 'lkm' arvoa.

if lkm > len(invariantti_massa):
    print('''\n Datan määrä on valittu suuremmaksi kuin mitä dataa on saatavilla, joten pyytämääsi
 histogrammia ei voitu piirtää. Maksimiarvo datan määrälle on %i.''' % len(invariantti_massa))

else:    
    for f in range(lkm):
        M = invariantti_massa[f]
        valitut.append(M)
    print('\n Valitsit koko datasta %i arvoa invariantille massalle.' %(lkm))

# Käytetään numpy-paketin histogram-funktiota, ja luodaan histogrammi valituista invarianteista massoista.
# Nimetään histogrammi "histogrammi1":ksi.

histogrammi1 = np.histogram(valitut, bins=120, range=(60,120))


# Mitä tapahtuu, jos muutat yllä olevia binien ja rangen arvoja?


# Valmistellaan histogrammin piirtoa.
# Määritetään histogrammin palkkien leveys sekä histogrammin keskikohta.

hist1, bins1 = histogrammi1
width1 = 1.0*(bins1[1] - bins1[0])
center1 = (bins1[:-1] + bins1[1:])/2

# Piirretään histogrammi matplotlib.pyplotin avulla.

plt.bar(center1, hist1, align='center', width=width1)

# Nimetään x- ja y-akselit sekä annetaan histogrammille otsikko.

plt.xlabel('Invariantti massa [GeV/c²]')
plt.ylabel('Tapahtumien lukumäärä pylvästä kohti', fontsize=10)
plt.title('Invariantin massan histogrammi kahdelle myonille\n', fontsize=15)

# Asetetaan y-akseli vakioksi välille 0-800.

axes = plt.gca()
axes.set_ylim([0,800])


# Kokeile muuttaa y-akselin näytettävää väliä. Mitä tapahtuu, jos et määrittele sitä lainkaan?
# Voit kokeilla kommentoimalla koodirivin #-merkin avulla.

# Miten saat muutettua x-akselin väliä? Mistä nykyinen väli tulee?


# Nollataan lopuksi valitut arvot uutta valintaa varten.

valitut = []
```

Kokeilepa muuttaa yllä olevaa koodia niin, että histogrammin piirtoalue (range) ei näytä kuin halutun piikin. Otsikotkin saattaisivat kaivata pientä korjaamista.

Tehdään sitten toinenkin jako: eritellään 'data' kahteen uuteen muuttujaan, joissa erotellaan mittatilanteet sen mukaan, onko havaittu myonipari ollut korkea- vai matalaenerginen.

```python
# Asetetaan jako 150 GeVin kohdalle (energiassa, ei massassa).
# Löydätkö 'data'-muuttujasta sarakkeet, joihin tässä viitataan?

uusiSettiKorkeaE = data[data.E1+data.E2>150]
uusiSettiMatalaE = data[data.E1+data.E2<150]
```

Sitten piirretään nämä molemmat yhtä aikaa näkyviin. Tai tarkemmin sanoen niiden M-sarakkeet.

```python
# Otsikot

plt.xlabel('Invariantti massa [GeV/c$^2$]')
plt.ylabel('Tapahtumien lukumäärä')
plt.title('Kahden myonin invariantti massa, vertailu energian perusteella \n')

# Itse kuvaajan piirto

plt.hist(uusiSettiMatalaE ['M'], bins=50, range=(80,100), alpha=0.5, label='Matala E')
plt.hist(uusiSettiKorkeaE ['M'], bins=50, range=(80,100), alpha=0.5, label='Korkea E')

# Piirretään selite. Voit kokeilla sille eri paikkoja, kuten 'lower left' jne.

plt.legend (loc='upper right')
plt.show()
```

Mitä luulet, ovatko kuvassa näkyvät Matala E ja Korkea E tehty samanarvoisista aineistoista? Nythän kuva antaisi kenties ymmärtää, että matalalla energialla tapahtuisi enemmän kiinnostavia löytöjä Z-bosonin massan (~91 Gev/c$^2$) kieppeillä, mutta histogrammia tai pylväsdiagrammia käyttäessä aineiston koolla on suuri merkitys.
